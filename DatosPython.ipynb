## Normalizado

```python
with open('data/carpetas-de-investigacion-pgj-de-la-ciudad-de-mexico_2019_Febrero.json') as file_:
    d = json.load(file_)
```

```python
lista_delitos = []
for delito in df['fields.delito'].unique():
    lista_delitos.append([delito,len(df[df['fields.delito'] == delito])])
df_delitos = pd.DataFrame(lista_delitos,columns = ['Delito' , 'Eventos'])
```

```python
df_delitos.sort_values(by='Eventos',ascending=False,inplace=True)
```

```python
df_delitos
```

```python
df_delitos.set_index(np.arange(len(lista_delitos))+1)
```

```python
df_delitos.plot(x='Delito',y='Eventos',loglog=True)
```
